```python
#Import libraries
from matplotlib_venn import venn2
import pandas as pd
import matplotlib.pyplot as plt

```

```python
tissues = ['Lymph node', 'Tonsil']
a = pd.read_csv(f'{tissues[0]}_mean_abundance_added.csv', sep='/')
b = pd.read_csv(f'{tissues[1]}_mean_abundance_added.csv', sep='/')
cutoff_a= a[a['importance']> 0].shape[0]+1
cutoff_b = b[b['importance']> 0].shape[0]+1
a_df = a[['feature_name', 'importance']].iloc[:cutoff_a, :]
a = set(a_df['feature_name'].tolist())

b_df = b[['feature_name', 'importance']].iloc[:cutoff_b, :]
b = set(b_df['feature_name'].tolist())

plt.figure(figsize=(10,12))
v=venn2([a, b],set_labels=(tissues))
plt.title('Overlap between features', fontsize=14)
plt.savefig('venn2_lymphnode', dpi=1000)

plt.show()
```

```python
shared_proteins = a.intersection(b)
only_a = a_df[~a_df['feature_name'].isin(shared_proteins)]['importance'].values.tolist()
only_b = b_df[~b_df['feature_name'].isin(shared_proteins)]['importance'].values.tolist()
shared_a= a_df[a_df['feature_name'].isin(shared_proteins)]['importance'].values.tolist()
shared_b=b_df[b_df['feature_name'].isin(shared_proteins)]['importance'].values.tolist()
plot_data={'unique lymph node': only_a, 'unique tonsil': only_b, 'shared lymph node':shared_a, 'shared tonsil':shared_b}

# sort keys and values together
sorted_keys, sorted_vals = zip(*sorted(plot_data.items(), key=op.itemgetter(1)))
sns.set(context='notebook', style='whitegrid')
#sns.axlabel(xlabel="Groups", ylabel="Y-Axis", fontsize=16)
sns.boxplot(data=sorted_vals, width=.18).set(ylabel='feature importance')
sorted_keys=('unique lymph node', 'unique tonsil',  'shared lymph node','shared tonsil')
# category labels
plt.xticks(plt.xticks()[0], sorted_keys, rotation=60)
plt.savefig('FI_lymph_tonsil.png', dpi=1000, bbox_inches='tight')
plt.show()
```
